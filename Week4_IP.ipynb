3. Cleaning the dataset

3.1 Consistency


```
# We can start by deleting the car column as it is a duplicate of the Bluecar counter
car_1 = Car_df.drop(columns=['Cars'])
car_1
```

3.2 Validity

```
# We drop scheduled at and Displayed comment column which is not important for our data analysis
car_2 = car_1.drop(columns=['Displayed comment', 'Scheduled at'])
car_2
```

```
#we delete row that are not useful in our dataset
# Stations that have no resources may not be useful in our data.
# They include:
# where Kind == CENTER
# Where Rental Status == future or Where Status = Scheduled
car_new = car_2.drop(car_2[(car_2['Kind'] == 'CENTER') & (car_2['Rental status'] == 'future')].index)
car_new
```

```
# delete all rows with column 'Bluecar counter' is not in the range 0-7
notrange = car_new[ (car_new['Bluecar counter'] < 0) & (car_new['Bluecar counter'] > 7) ].index
car_3 = car_new.drop(notrange, inplace=True)
print(car_3)
```

```
# delete all rows with column 'Utilib counter' is not in the range 0-4
notrange = car_new[ (car_2['Utilib counter'] < 0) & (car_2['Bluecar counter'] > 4) ].index
car_2.drop(notrange , inplace=True)
car_2.shape
```
