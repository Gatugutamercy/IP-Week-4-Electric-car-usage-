```
# Importing the Libraries
import pandas as pd
import numpy as np

```

1. Loading the dataset


```
# loading the Electric car dataset

Car_df = pd.read_csv('http://bit.ly/autolib_dataset')
Car_df
```

2. Previewing the dataset

```
#Getting information abou the data
Car_df.info()
```

```
#Understanding the dataset's shape
Car_df.shape
```

3. Cleaning the dataset

3.1 Uniformity

```
#  We can add _ between the column names with 2 words for easier analysis
Car_df.columns = [c.replace(' ', '_') for c in Car_df.columns]
Car_df
```

3.2 Consistency


```
#  We can add _ between the column names with 2 words for easier analysis
Car_df.columns = [c.replace(' ', '_') for c in Car_df.columns]
Car_df
```

```
Car_df['Date']=pd.to_datetime(Car_df[['year','month','day', 'hour', 'minute']])
Car_df
```

```
# We can start by deleting the car column as it is a duplicate of the Bluecar counter and Year, and month data was collected on April, 2018
car_1 = Car_df.drop(columns=['Cars', 'year', 'month', 'day', 'hour', 'minute'])
car_1
```

3.3 Validity

```
# We drop scheduled at and Displayed comment column which is not important for our data analysis
car_2 = car_1.drop(columns=['Displayed_comment', 'Scheduled_at'])
car_2.shape
```

```
#we delete row that are not useful in our dataset
# Stations that have no resources may not be useful in our data.
# we will drop the first 2 which will cater for the rest
# They include:
# where Kind == CENTER
# Where Rental Status == future 
# Where Status = Scheduled
# Where Charging Status == future
# Where Subscription Status == future

car_3 = car_2[car_2.Kind != 'CENTER']
car_3.shape
```

```
car_4 = car_3[car_3.Status != 'scheduled']
car_4.shape
```

3.4 Accuracy

```
# We proceed to chwck columns that are not within the specified range ie
# Bluecar_counter (0-7)
# Utilib_Counter (0-5)
# Utilib_1.4_counter
# Charge_slots (0-3)
((car_4['Bluecar_counter'] > 7) & (car_4['Bluecar_counter'] < 0)).any()
# all are within the range
```

```
((car_4['Utilib_counter'] > 4) & (car_4['Utilib_counter'] < 0)).any()
#All are within the range
```

```
((car_4['Utilib_1.4_counter'] > 5) & (car_4['Utilib_1.4_counter'] < 0)).any()
# All are within the range
```

```
((car_4['Charge_Slots'] > 5) & (car_4['Charge_Slots'] < 0)).any()
#All are within the range
```

4. Data Analysis

```
# First we selecting all tye rows with the city as Paris and name the data df
df = car_4[car_4.City == 'Paris']
df
```

```
# We need to arrange the data in ascending order on the date column
df_1 = df.sort_values('Date',ascending=True)
df_1.head()
```

```
df_2 = df_1.sort_values(['ID', 'Date'], ascending=[False, False])
df_2
```

```
# finsing the difference between consequtive rows in Blue cars to identify the hour where cars were mostly taken from the stations
df_2["Bluecar_diff"] = df_2["Bluecar_counter"].diff(1)
df_2
# the - value in Bluecar_diff means that the car was taken from the station, while + value means that the car was brought back to the station.
```

```
# Now we look at the time when most blue cars were brought back to the stations
df_2.nsmallest(3, 'Bluecar_diff')
#this will provide the times that were most popular for bringing back the blue cars
```

```
# Now we look at the most popular hour for picking up blue cars, by checking when most cars were taken from the Autolib stations
df_2.nlargest(3, 'Bluecar_diff')
#this will provide the hours that were most popular for picking the blue cars
```
